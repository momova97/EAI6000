Basic example of each algorithm using Python and some popular libraries such as scikit-learn and matplotlib.

Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate some data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 3 + 4 * X + np.random.randn(100, 1)

# Fit the model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Make predictions
X_new = np.linspace(0, 2, 100).reshape(100, 1)
y_pred = lin_reg.predict(X_new)

# Plot the results
plt.scatter(X, y)
plt.plot(X_new, y_pred, 'r')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

Logistic Regression:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate some data
np.random.seed(42)
X, y = make_classification(n_samples=100, n_features=1, n_informative=1, n_redundant=0, n_clusters_per_class=1)

# Fit the model
log_reg = LogisticRegression()
log_reg.fit(X, y)

# Make predictions
X_test = np.linspace(-3, 3, 300)
y_prob = log_reg.predict_proba(X_test.reshape(-1, 1))[:, 1]

# Plot the results
plt.scatter(X, y)
plt.plot(X_test, y_prob, 'r')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

Polynomial Regression:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# Generate some data
np.random.seed(42)
X = 2 * np.random.rand(100, 2)
y = 3 + X[:, 0]**2 + X[:, 1]**2 + np.random.randn(100)

# Fit the model
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

# Make predictions
X1_new = np.linspace(0, 2, 100)
X2_new = np.linspace(0, 2, 100)
X1_new, X2_new = np.meshgrid(X1_new, X2_new)
X_new = np.c_[X1_new.ravel(), X2_new.ravel()]
X_new_poly = poly_features.transform(X_new)
y_pred = lin_reg.predict(X_new_poly)

# Plot the results
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y)
ax.plot_surface(X1_new, X2_new, y_pred.reshape(X1_new.shape), alpha=0.5, color='r')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
plt.show()
```

k-Nearest Neighbors (kNN):

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier

# Generate some data
np.random.seed(49)
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1)

# Fit the model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Make predictions
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the results
plt.contourf(xx, yy, Z, alpha=0.6)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

Support Vector Machine (SVM):

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC

# Generate some data
np.random.seed(49)
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1)

# Fit the model
svm = SVC(kernel='linear')
svm.fit(X, y)

# Make predictions
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the results
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

Principal Component Analysis (PCA):


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Generate some data
np.random.seed(49)
X = np.random.randn(100, 2)

# Fit the model
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X_pca, np.zeros(X_pca.shape[0]), c='r')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

k-Means Clustering:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate some data
np.random.seed(40)
X, y = make_blobs(n_samples=100, centers=3, cluster_std=1.0)

# Fit the model
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
